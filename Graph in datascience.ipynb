# FactorPlot

```python
data=sns.load_dataset("exercise")
sns.catplot(x='time',y='pulse',hue='kind',kind='violin',col='diet',data= data)
```

# 1.Pulse increases with activity intensity
Rest < Walking < Running in pulse rate across all time intervals and diets.
Running consistently shows the highest pulse.
2.Pulse increases with time
For each activity type, pulse generally rises from 1 min â†’ 15 min â†’ 30 min.
The increase is most pronounced for running.
3.Diet effect
Low-fat diet group generally shows slightly lower pulse values compared to the no-fat group for the same activity and time.
The difference is subtle but consistent.
4.Variability (distribution width)
Running has the widest violin shapes, indicating higher variability in pulse.
Rest shows tighter distributions (more stable pulse).
5.Peak pulse
The highest pulse values appear for running at 30 min, especially in the no-fat diet group.
The plot compares pulse rate across:
Time durations: 1 min, 15 min, 30 min
Activity types (kind): Rest, Walking, Running
Diet groups: No fat vs Low fat (separate panels)
