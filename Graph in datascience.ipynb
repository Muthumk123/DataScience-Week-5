```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("Preplacementdata.csv")
```

# number of rows and colums code

```python
print(" Number of rows in dataset is:" ,df.shape[0])
print(" Number of columns in dataset is:" ,df.shape[1])
```

```python
sns.distplot(df['ssc_p'],hist=True,kde=True,kde_kws={'color':'Red'})
```

# Bivarite Analiys :Distribution Plot

```python
sns.jointplot(x='hsc_p',y='salary',data=df)
plt.show()
```

```python
sns.jointplot(x='hsc_p',y='salary',data=df ,kind='hex')
plt.show()
```

```python
sns.jointplot(x='hsc_p',y='salary',data=df ,kind='kde')
plt.show()
```

# Pair plot

```python
sns.pairplot(df,hue='gender',diag_kind='kde',kind='scatter',palette='husl')
plt.show()
```

# for catagorical polt

```python
sns.stripplot(x='ssc_b',y='ssc_p',data=df)
plt.show()
```

```python
sns.stripplot(x='ssc_b',y='ssc_p',data=df,jitter=True)
plt.show()
```

```python
sns.swarmplot(x='gender',y='ssc_p',data=df)
plt.show()
```

# Distribution observation

```python
sns.violinplot(x='gender',y='ssc_p',data=df)
plt.show()
```

```python
sns.violinplot(x='specialisation',y='salary',data=df,hue='gender')
plt.show()
```

```python
sns.boxplot(x='ssc_b',y='ssc_p',data=df,hue='gender')
plt.show()
```
