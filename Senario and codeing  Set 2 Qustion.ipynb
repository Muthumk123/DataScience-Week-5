# Scenario 1: 
# Flight Delay Analysis
# An airline tracks flight delays (in minutes) for 20 flights. Analyze the flight delays to calculate percentiles, detect outliers, and evaluate the overall distribution.

```python
import numpy as np
delays = [5, 10, 12, 15, 18, 20, 22, 25, 30, 35,
          40, 45, 50, 55, 60, 65, 70, 90, 120, 180]

Q1 = np.percentile(delays, 25)
Median = np.percentile(delays, 50)
Q3 = np.percentile(delays, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in delays if x < lower_bound or x > upper_bound]
mean_delay = np.mean(delays)
std_delay = np.std(delays)
print("MD:",mean_delay,"\n" "STD:", std_delay)
print("IQR:",IQR,"\n","Q1:",Q1,"\n","Median:",Median,"\n","Q3:",  Q3)
print ( "LB:" , lower_bound,"\n","UB:",upper_bound,"\n","outerline:",outliers)


```

# Scenario 2: 
# Employee Salary Analysis
# A company wants to analyze the salary distribution of its employees to understand the central tendency and determine whether the data is skewed.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew

# Sample salary data (replace with your dataset)
salary = [25000, 30000, 32000, 35000, 40000, 45000, 
          50000, 60000, 75000, 120000, 200000]

# Convert to DataFrame
df = pd.DataFrame({'Salary': salary})

# Central Tendency
mean_salary = df['Salary'].mean()
median_salary = df['Salary'].median()
mode_salary = df['Salary'].mode()[0]

print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)

# Skewness
salary_skewness = skew(df['Salary'])
print("Skewness:", salary_skewness)

# Histogram
plt.hist(df['Salary'], bins=8)
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Salary Distribution")
plt.show()

# Boxplot
plt.boxplot(df['Salary'], vert=False)
plt.title("Salary Boxplot")
plt.show()
```

# Scenario 5: 
# Clinical Trial for Diabetes Medication
# A pharmaceutical company conducted a clinical trial with two groups: one receiving medication and the other a placebo. 
# Perform a hypothesis test to determine the effectiveness of the medication.

```python
import pandas as pd
from scipy.stats import ttest_ind
data = {
    'Group': ['Treatment']*10 + ['Control']*10,
    'Blood_Pressure': [
        120, 115, 118, 123, 122, 119, 124, 117, 116, 121,   # Medication
        130, 135, 140, 138, 142, 136, 139, 134, 137, 141    # Placebo
    ]
}
df = pd.DataFrame(data)
df
treatment = df[df['Group'] == 'Treatment']['Blood_Pressure']
control = df[df['Group'] == 'Control']['Blood_Pressure']
t_stat, p_value = ttest_ind(treatment, control)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis: The medication is effective.")
else:
    print("Fail to Reject the Null Hypothesis: No significant effect.")
```

# Question 1 :
# An airline tracks flight delays (in minutes) for 20 flights. Analyze the flight delays to calculate percentiles, detect outliers, and evaluate the overall distribution.
# DataSet:
# delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50, 
#           120, 80, 10, 75, 90, 200, 55, 40, 70, 150]
# Expected Output:

# Percentiles (10th, 25th, 50th, 75th, 90th): [ 14.5   28.75  52.5   82.5  123.  ]
# IQR: 53.75
# Outliers: [200]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Delivery delay data (in minutes)
delays = [15, 30, 45, 20, 25, 100, 5, 60, 35, 50,
          120, 80, 10, 75, 90, 200, 55, 40, 70, 150]

# Create DataFrame
df = pd.DataFrame(delays, columns=["Delay"])

# Mean and Standard Deviation
mean_delay = np.mean(delays)
std_delay = np.std(delays)

# Percentiles
Q1 = np.percentile(delays, 25)
Median = np.percentile(delays, 50)
Q3 = np.percentile(delays, 75)
P10 = np.percentile(delays, 10)
P90 = np.percentile(delays, 90)

# IQR
IQR = Q3 - Q1

# Outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = df[(df["Delay"] < lower_bound) | (df["Delay"] > upper_bound)]

# Print results
print("Mean Delay:", mean_delay)
print("Standard Deviation:", std_delay)
print("\nQ1:", Q1)
print("Median:", Median)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("\nOutliers:")
print(outliers)

# Boxplot
plt.figure(figsize=(8, 4))
plt.boxplot(df["Delay"], vert=False)
plt.title("Flight Delay Boxplot")
plt.xlabel("Delay (Minutes)")
plt.show()
```

# Question 2 :
# A company wants to analyze the salary distribution of its employees to understand the central tendency and determine whether the data is skewed.
# DataSet:
# salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
#             60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
# Expected Output:
# Mean Salary: 56950.0
# Median Salary: 57500.0
# Mode Salary: 30000
# The data is Left Skewed (Negative Skew)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
salaries = [30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000]
df=pd.DataFrame(salaries ,columns=['Salary'])
mean_salary = np.mean(salaries)
median_salary = np.median(salaries)
mode_salary = stats.mode(salaries, keepdims=True).mode[0]
print("Mean Salary:",mean_salary)
print("Median Salary:",median_salary)
print("Mode Salary:",mode_salary)
# Visualization using Histogram
plt.figure(figsize=(12,5))
# Histogram
plt.subplot(1,2,2)
sns.histplot(salaries, kde=True)
plt.title('Histogram of Employee Salaries')
plt.tight_layout()
plt.show()
```

# Question 4:
# A pharmaceutical company conducted a clinical trial with two groups: one receiving medication and the other a placebo. How do you perform a hypothesis test to determine the effectiveness of the medication?
# Dataset:
# medication_group = [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
# placebo_group = [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]
# Expected Output:
# T-Statistic: -9.201427649220966
# P-Value: 3.163912817600812e-08
# Reject the null hypothesis: The medication is effective.

```python
import pandas as pd
from scipy.stats import ttest_ind

medication_group= [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
placebo_group= [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]
t_stat, p_value = ttest_ind(medication_group, placebo_group)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis: The medication is effective.")
else:
    print("Fail to Reject the Null Hypothesis: No significant effect.")
```
