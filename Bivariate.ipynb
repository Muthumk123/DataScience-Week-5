```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def vif_calcu(X):
    #calculating VIF
    vif=pd.DataFrame() # Creating table
    vif["variables"]=X.columns # For given input X in column name act as varabiles to create columns
    vif["VIF"]=[variance_inflation_factor(X.values,i)for i in range(X.shape[1])] # hear fist for loop will execute in I will take X values and in
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    vif["Tolerance"] = 1 / vif["VIF"]
    return (vif)
    
```

# Statement: Test whether 55C p of male and HSC P of Male is Same?

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
Commerce=dataset[dataset['hsc_s']=='Commerce']["mba_p"]
Science=dataset[dataset['hsc_s']=='Science']["mba_p"]
ttest_ind(Commerce,Science)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('salary ~ C(degree_p) + C(mba_p) + C(degree_p):C(mba_p)',data=dataset).fit()
anova_table = sm.stats.anova_lm(model,typ=2)
print(anova_table)
```
